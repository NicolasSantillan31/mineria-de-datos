En este trabajo se aplican conceptos de minería de datos para analizar un conjunto de datos de clientes y observar patrones de comportamiento.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

datos = {
    "Cliente": [1, 2, 3, 4, 5],
    "Edad": [25, 40, 32, 22, 35],
    "Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
    "Consumo_GB": [5, 50, 10, 7, 45],
    "Reclamos": [1, 0, 2, 3, 0],
    "Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}

df = pd.DataFrame(datos)

print(df)
```

```python
# Histograma
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.title("Consumo de datos")
plt.show()

# Barras
sns.countplot(x=df["Estado_Cuenta"])
plt.title("Estado de cuenta")
plt.show()
```

Se observa que los clientes con mayor consumo de datos suelen estar en estado "Pagado", mientras que los clientes morosos presentan más reclamos. Esto puede indicar una relación entre los reclamos y la morosidad.
