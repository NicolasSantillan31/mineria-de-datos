```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

df
```

# Análisis de clientes
En este trabajo se analizan datos de clientes de una tienda en línea.
La idea es usar distintos gráficos para observar patrones entre edad, ingresos y compras.

```python
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()
```

```python
sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.show()
```

```python
sns.scatterplot(x=df["Ingresos"], y=df["Compras"])
plt.title("Relación entre Ingresos y Compras")
plt.show()
```

```python
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()
```

```python
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True)
plt.show()
```

```python
sns.violinplot(x="Categoría", y="Compras", data=df)
plt.show()
```

## Análisis exploratorio

Primero se hicieron algunos gráficos básicos.

El histograma muestra cómo se distribuyen las edades de los clientes.

El boxplot permite ver si hay valores muy diferentes en la cantidad de compras.

En el gráfico de dispersión se puede observar que, en general, cuando los ingresos son mayores, también aumentan las compras.

## Visualización avanzada

Después se usaron gráficos más completos.

El pairplot ayuda a ver varias relaciones entre las variables numéricas al mismo tiempo.

El heatmap muestra la correlación entre las variables. En este caso se observa una relación positiva entre ingresos y compras.

El violin plot permite comparar las compras según cada categoría.
